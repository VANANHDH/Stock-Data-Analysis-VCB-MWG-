## Portfolio Optimization

```python
# Concat VCB and MWG
portfolio = pd.DataFrame()
portfolio['VCB'] = VCB['Close']
portfolio['MWG'] = MWG['Close']
```

```python
portfolio
```

```python
# Normalize to start at the same price: close price/start-close-price *100
ret = portfolio/portfolio.iloc[0]*100
```

```python
ret
```

```python
ret.plot(figsize=(12,8));
```

```python
### Calculate log return, covariance, and correlation
# daily log return: log of (close price / close price 1 index before)
log_ret = np.log(portfolio/portfolio.shift(1))
log_ret
```

```python
# Annualized log return: mean of daily log return * number of transaction day
log_ret.mean()*252
```

```python
# Annualized covariance (hard to determine the interrelation)
log_ret.cov()*252
```

```python
# Correlation (from -1 to 1)
log_ret.corr()
```

### Weight

```python
# Weights (tỷ trọng)
num_stocks = 2 # VCB and MWG
```

```python
# create random weights (make sure the sum = 1)
weight = np.random.random(num_stocks)
weight = weight/ np.sum(weight)
weight[0] + weight[1] == 1
```

```python
weight
```

```python
# Expected annualized porfolio return
np.sum(weight * log_ret.mean())*252
```

```python
# Expected portfolio variance (dot product of 2 weights)
port_var = np.dot(weight.T, np.dot(log_ret.cov()*252, weight))
port_var
```

```python
# Expected portfolio volatility (standard deviation = square root of portfolio variance)
port_vol = np.sqrt(port_var)
port_vol
```

## Run 1000 simulation with different weights to choose the best efficient portfolio

```python
# creat 2 empty lists for portfolio return and portfolio volatility
port_ret = []
port_vol = []

# create 1000 random weights and calculate portfolio return and portfolio volatility
for i in range(1000):
    weight = np.random.random(num_stocks)
    weight /= np.sum(weight)
    port_ret.append(np.sum(weight * log_ret.mean())*252)
    port_vol.append(np.sqrt(np.dot(weight.T, np.dot(log_ret.cov()*252, weight))))

# convert list to array for plotting efficient frontier
port_ret = np.array(port_ret)
port_vol = np.array(port_vol)
```

### Plot efficient frontier

```python
# Create a portfolio with 2 columns: return and volatility
portfolio = pd.DataFrame({'Return': port_ret, 'Volatility': port_vol})
```

```python
portfolio
```

```python
portfolio.plot(x='Volatility', y='Return', kind='scatter', figsize=(12,8))
plt.title('Efficient Frontier between MWG and VCB')
plt.xlabel('Volatility', fontsize=15)
plt.ylabel('Return', fontsize=15)
```
